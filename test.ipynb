```python
from Calculator.Structure import Structure
import matplotlib.pyplot as plt
struct = Structure()
struct.layers = "Ni"
struct.parseLayers(delimeter=" ")
print(struct.list_layers)
model = struct.initModel()
model.searchKD()

time = [int(i) for i in range(7200)]
sorb = []
desorb = []
for t in time:
    model.searchTime(time=t)
    sorb.append(model.searchSorb())
    desorb.append(model.searchDesorb())

plt.plot(time, sorb, label="Sorb")
plt.plot(time, desorb, label="Desorb")
plt.xlabel("Time, s")
plt.ylabel('Sorb/Desorb')
plt.legend()
plt.grid()
```
